### ML class 1 

```python
# Label Encoding

# Cloudy --> 0
# Rainy --> 1
# Sunny --> 2
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labels = ["Cloudy", "Rainy", "Sunny"]
```

```python
le = LabelEncoder()
```

```python
le.fit(labels)
```

```python
print('\nClass mapping\n')
for i, item in enumerate(le.classes_): 
    print(item, '-->', i)
```

```python
label2 = ["Sunny", "Sunny", "Rainy", "Cloudy"]
```

```python
encoded_labels = le.transform(label2)
```

```python
print("Encoded Labels", list(encoded_labels))
```

```python
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept = True)
```

```python
x.shape
```

```python
# set the dimension
X = x[:, np.newaxis]
X.shape
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
```

```python
y_pred = model.predict(Xfit)
```

```python
plt.scatter(x,y)
plt.plot(Xfit, y_pred, color='red');
```

```python
from sklearn.metrics import r2_score
y_fit = model.predict(X)
r2_score(y, y_fit)
```

### Back to Iris

```python
iris['target'] = le.fit_transform(iris['species'])
iris.head()
```

```python
X_iris = iris.drop(['species', 'target'], axis=1)
y_iris = iris['target']
```

```python
y_iris = y_iris[:, np.newaxis]
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, test_size=0.2, random_state=1)
```

```python
Xtrain.shape, ytrain.shape
```

```python
Xtest.shape, ytest.shape
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(Xtrain, ytrain)
yiris_pred = gnb.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, yiris_pred)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_iris)
x2d = pca.transform(X_iris)
iris['PCA1'] = x2d[:, 0]
iris['PCA2'] = x2d[:, 1]
sns.lmplot(data=iris, x="PCA1", y='PCA2', hue='species', fit_reg=False);
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3, covariance_type='full')
gmm.fit(X_iris)
y_gmm = gmm.predict(X_iris)
```

```python
iris['cluster'] = y_gmm
iris.head()
```

```python
sns.lmplot(data=iris, x='PCA1', y='PCA2', col='cluster', hue='species', fit_reg=False);
```
